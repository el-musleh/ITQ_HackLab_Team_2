```python
from SCSCtrl import TTLServo
import time

print("Servo controller loaded.")
```

```python
# Claw servo ID
CLAW_SERVO_ID = 4

# Claw angles
CLAW_OPEN = -10
CLAW_CLOSED = -75

# Arm positions
ARM_HOME_X = 130
ARM_HOME_Y = 20

ARM_DOWN_X = 150
ARM_DOWN_Y = -45

ARM_UP_X = 120
ARM_UP_Y = 45
```

```python
TTLServo.servoAngleCtrl(
    CLAW_SERVO_ID,
    CLAW_OPEN,
    1,
    150
)

print("Claw Open")
```

```python
TTLServo.servoAngleCtrl(
    CLAW_SERVO_ID,
    CLAW_CLOSED,
    1,
    150
)

print("Claw Closed")
```

```python
TTLServo.xyInputSmooth(
    ARM_HOME_X,
    ARM_HOME_Y,
    0.8
)

print("Arm Home")
```

```python
TTLServo.xyInputSmooth(
    ARM_DOWN_X,
    ARM_DOWN_Y,
    0.8
)

print("Arm Down")
```

```python
TTLServo.xyInputSmooth(
    ARM_UP_X,
    ARM_UP_Y,
    0.8
)

print("Arm Up")
```

### Full pick-up Sequence

```python
print("Starting pickup test")

# Open claw
TTLServo.servoAngleCtrl(4, CLAW_OPEN, 1, 150)
time.sleep(1)

# Lower arm
TTLServo.xyInputSmooth(
    ARM_DOWN_X,
    ARM_DOWN_Y,
    0.8
)
time.sleep(2)

# Close claw
TTLServo.servoAngleCtrl(4, CLAW_CLOSED, 1, 150)
time.sleep(1)

# Lift arm
TTLServo.xyInputSmooth(
    ARM_UP_X,
    ARM_UP_Y,
    0.8
)
time.sleep(2)

print("Pickup sequence complete")
```

### Drop Sequence

```python
print("Starting drop test")

# Move above box
TTLServo.xyInputSmooth(
    ARM_DOWN_X,
    ARM_DOWN_Y,
    0.8
)
time.sleep(2)

# Release object
TTLServo.servoAngleCtrl(
    CLAW_SERVO_ID,
    CLAW_OPEN,
    1,
    150
)
time.sleep(1)

# Return home
TTLServo.xyInputSmooth(
    ARM_HOME_X,
    ARM_HOME_Y,
    0.8
)

print("Drop complete")
```

### Automatic Repeating
Calibration

```python
for _ in range(5):  # patched infinite loop

    print("OPEN")
    TTLServo.servoAngleCtrl(4, CLAW_OPEN, 1, 150)
    time.sleep(2)

    print("CLOSE")
    TTLServo.servoAngleCtrl(4, CLAW_CLOSED, 1, 150)
    time.sleep(2)
```
